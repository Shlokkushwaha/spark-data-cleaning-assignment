```python
#MapReduce stores intermediate results on disk after every operation, which makes it slow. It also lacks support for iterative processing and real-time analytics. Apache Spark overcomes these limitations using in-memory computation, which makes processing much faster, and it supports iterative algorithms, machine learning, and real-time data processing efficiently.
```

```python
#Spark stores intermediate data in memory (RAM) instead of writing it to disk. This reduces disk I/O and speeds up iterative algorithms like machine learning models and graph processing because data can be reused without repeated disk access.
```

```python
df = spark.table("workspace.default.spark_dataset_1")
df.display()
```

```python
df_clean = df.dropDuplicates(["user_id", "transaction_date"])
df_clean.display()
```

```python
from pyspark.sql.functions import avg
result = (df.filter(df.region == "West") \
  .groupby("product_category") \
  .agg(avg("sale_amount").alias("avg_sale_amount")))

result.display()
```

```python
#na.drop() removes rows with null values
#na.fill() replaces null values with a specified value
```

```python
df.na.drop().display()
```

```python
df_filled = df.na.fill(
  {"status":"unknown"}
)
df_filled.display()
```

```python
from pyspark.sql.functions import count
result = (df.groupby("city").agg(count("*").alias("total_count")).filter("total_count > 100"))
result.display()
```

```python
#Spark DataFrames are immutable, meaning once created they cannot be changed. Any transformation like drop or rename creates a new DataFrame. This ensures data safety, lineage tracking, and fault tolerance.
```

```python
result = df.filter(
  (df.age >= 18) & (df.age <= 30) & (df.subscription == "Premium")
)
display(result)
```

```python
#Null values can distort results of sum, average, and count operations. Handling nulls ensures accurate and meaningful aggregation results.
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType

df_new = (
    df.withColumn(
        "event_time",
        col("raw_timestamp").cast(TimestampType())
    )
    .drop("raw_timestamp")
)

display(df_new)
```

```python
#Shuffle is the process of redistributing data across partitions during operations like groupBy or join. It is called a wide transformation because data moves across multiple nodes, which is expensive and slow.
```

```python
from pyspark.sql.functions import col
df = spark.table("workspace.default.spark_dataset_1")
clean_df = df.filter(
    col("email").isNotNull()
).filter(
    col("username") != ""
)

display(clean_df)
```

```python
from pyspark.sql.functions import min, max, mean

result = df.agg(
    min("price").alias("min_price"),
    max("price").alias("max_price"),
    mean("price").alias("avg_price")
)

display(result)
```

```python
#Using inferSchema=True with messy data can lead to incorrect data types being assigned. For example, inconsistent date formats or mixed types may be wrongly interpreted, causing errors in analysis.
```

```python
from pyspark.sql.functions import sum

result = (
    df.dropDuplicates()
      .na.fill({"price": 0})
      .groupBy("store_id")
      .agg(
          sum("price").alias("total_revenue")
      )
)

display(result)
```
